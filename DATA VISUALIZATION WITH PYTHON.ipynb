# SCATTER PLOT

```python
#scatter plot
#with year and income
plt.scatter(ds['year'],ds['income'],color="orange")

#for title
plt.title("INCOME FOR EACH YEAR")

#for label
plt.xlabel("YEAR")
plt.ylabel("INCOME")

#for show
plt.show()
```

# LINE CHART

```python
#plotting a line chart
plt.plot(ds["year"],color="yellow")
plt.plot(ds["expenditure"],color="green")

#title
plt.title("YEAR-WISE EXPENDITURE")

#LABELS
plt.xlabel("YEAR")
plt.ylabel("EXPENDITURE")

#show
plt.show()
```

# BAR CHART

```python
#plotting a bar chart 
plt.bar(ds["year"],ds["own"],color="darkblue")

#title
plt.title("YEAR VS OWN")

#labels
plt.xlabel("YEAR")
plt.ylabel("OWN")

#show bar chart
plt.show()
```

# HISTOGRAM 

```python
#plotting an Histogram
plt.hist(ds["eqv_income"],color="gold",edgecolor="green")

#title
plt.title("YEAR")
#show fig
plt.show()
```

# HEAT MAP

```python
ds.corr()
```

```python
#plotting a figure size
plt.figure(figsize=(10,10))

#plotting heatmap
sns.heatmap(ds.corr(),cmap="viridis",annot=True)

#title
plt.title("CORRELATION")

#show
plt.show()
```
